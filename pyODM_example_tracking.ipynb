```python
images = glob.glob("*.JPG")
print(len(images), "images")
# images = images[190:300]
```

```python
n = Node('localhost', 3000)
```

```python
t = n.create_task(images, {'fast-orthophoto': True})
```

```python
info = t.info()
print("images: ", info.images_count)

for n in range(0,20):
    info = t.info()
    print(info.status)
    print(round(info.processing_time/3.6e+6,2), " hours")
    print(round(info.progress), " %")
    if info.progress == 100:
        print("This task completed in",round(info.processing_time/3.6e+6,2), "hours")
        break
    print("wait 1 min")
    time.sleep(1*60) # wait 10 minutes
```
